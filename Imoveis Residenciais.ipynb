## Imoveis residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
```

```python
 dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa Comercial',
 'Casa de Vila']
```

```python
residencial
```

```python
dados['Tipo'].isin(residencial)
```

```python
selecao = dados['Tipo'].isin(residencial)
```

```python
selecao
```

```python
dados_residencial = dados[selecao]
```

```python
dados_residencial.head(10)
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
dados_residencial.shape[0]
```

```python
dados_residencial.index = dados_residencial.shape[0]
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```
